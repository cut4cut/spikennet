```python
dataset_name = 'sleep-edfx'
files_path = '../data/raw/{}/'.format(dataset_name)

files_list = os.listdir(files_path)
choice_index = np.random.randint(len(files_list), size=1)[0]

dataset_name = files_list[choice_index]
```

```python
dataset = mne.io.read_raw_edf(files_path + dataset_name)
```

```python
dataset.ch_names
```

```python
data = dataset.get_data()
```
